(Problem #1) Face Blurring: You are expected to write an algorithm that blurs out people's faces in any given colored image without affecting the remaining parts of the image, if any face exists. this will pluring faces on vedio

**here I write code for image face blurring**

```python
# this code detect the face in image then blurrring it i try image for one face and image for multiple faces here 

import cv2

face_cascade = cv2.CascadeClassifier("C:/Users/roaas/Downloads/haarcascade_frontalface_default.xml")

img = cv2.imread("C:/Users/roaas/Downloads/girl.jpg")
#img = cv2.imread("C:/Users/roaas/Downloads/group.jpg")

detection = face_cascade.detectMultiScale(img, 1.1, 6)

#print(detection)

for (x, y, w, h) in detection:
    
    img[y:y+h, x:x+w] = cv2.GaussianBlur(img[y:y+h, x:x+w], (15,15), cv2.BORDER_DEFAULT)
    cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)
    cv2.imshow("output", img)

cv2.waitKey(0)
```

**here I write code for vedio face blurring**

```python
# this code detect face in video then blurring it i try short video here
import cv2
import numpy as np

def resize(img, new_width=500):
  
    hight, width, _ = img.shape

    # calculate the ratio
    ratio = hight/width
    new_hight = int(ratio*new_width)
    return cv2.resize(img, (new_width, new_hight))


face_cascade = cv2.CascadeClassifier("C:/Users/roaas/Downloads/haarcascade_frontalface_default.xml")

cap = cv2.VideoCapture("C:/Users/roaas/Downloads/example.mp4")

while True:
    _, frame = cap.read()
    frame = resize(frame)
    detection = face_cascade.detectMultiScale(frame, 1.1, 6)

    #print(detection)

    for (x, y, w, h) in detection:
        
        frame[y:y+h, x:x+w] = cv2.GaussianBlur(frame[y:y+h, x:x+w], (15,15), cv2.BORDER_DEFAULT)
        cv2.rectangle(frame, (x,y), (x+w, y+h), (255,0,0), 2)
        cv2.imshow("output", frame)
        
        if cv2.waitKey(1) == 27:
            break

cap.release()
cv2.destroyAllWindows()
```
